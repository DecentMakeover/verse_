Load imports

```python
import ipyvolume
import SimpleITK as sitk 
import os
import cv2
import seaborn as sns
%matplotlib inline
```

read a single image and label

```python
image_folder = '/media/ryan/shakira/miccai/data/images'
image = sitk.ReadImage(os.path.join(image_folder ,'verse005.nii.gz'))
image = sitk.GetArrayFromImage(image)
image.shape
```

check min and max values

```python
image.min(), image.max()
```

calculate histogram

```python
import matplotlib.pyplot as plt
import cv2

vals = image.ravel()
tentative_range = image.max() - image.min()
b, bins, patches = plt.hist(vals, tentative_range)
plt.xlim([image.min(),image.max()])
plt.show()
```
