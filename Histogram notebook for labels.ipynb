```python
import ipyvolume
import SimpleITK as sitk 
import os
import cv2
import seaborn as sns
%matplotlib inline
```

Read a single label

```python
mask_folder = '/media/ryan/shakira/miccai/data/masks'
mask = sitk.ReadImage(os.path.join(mask_folder ,'verse005_seg.nii.gz'))
mask = sitk.GetArrayFromImage(mask)
mask.shape
```

just check the min and mask values

```python
mask.min(), mask.max()
```

plot histogram

```python
import matplotlib.pyplot as plt
import cv2

vals = mask.ravel()
tentative_range = mask.max() - mask.min()
b, bins, patches = plt.hist(vals, tentative_range)
plt.xlim([mask.min(),mask.max()])
plt.show()
```
